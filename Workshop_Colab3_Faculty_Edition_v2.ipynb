## Instructor Background
Students often confuse visualization with understanding. Emphasize uncertainty, evidence, sample size, and reproducibility throughout the notebook.

## Common Misconceptions
1. A graph automatically explains a phenomenon.
2. Correlation implies causation.
3. AI interpretations are always correct.
4. Small datasets support broad conclusions.

```python
import pandas as pd
url="https://raw.githubusercontent.com/OpenNeuroDatasets/ds000247/main/participants.tsv"
df=pd.read_csv(url,sep="\t")
df.head()
```

```python
df=df[df["age"]!="n/a"].copy()
df["age"]=pd.to_numeric(df["age"])
df.describe(include="all")
```

## Activity 1: Exploratory Data Analysis
Students examine distributions, variables, missing information, and dataset limitations.

### Faculty Solution
Expected observations: small sample size, demographic metadata, limited variables, and restricted inference potential.

## Facilitation Notes
Ask students: What can be concluded? What cannot be concluded? Require evidence-based answers.

```python
import matplotlib.pyplot as plt
plt.hist(df["age"])
plt.title("Age Distribution")
plt.show()
```

## Worked Visualization Example
Use the age histogram to discuss variability, sample size, uncertainty, and overinterpretation risks.

## Interpretation Framework
Observe → Describe → Explain Limitations → Suggest Additional Data.
Students should not jump directly from visualization to causal conclusions.

## Activity 2 Faculty Solution
The visualization shows age variation among participants but does not support population-level conclusions.

## Strong Student Response
Connects observations to evidence, uncertainty, and dataset limitations.

## Average Student Response
Describes visible patterns but omits limitations.

## Weak Student Response
Claims broad scientific findings unsupported by the data.

## Neuroscience Discussion Case Study
Ask students which additional variables would be needed to investigate cognition, sleep, attention, or neurological conditions.

## Reproducibility Checkpoint
Document dataset source, variables examined, plots created, code used, and interpretation decisions.

## Discussion Prompts
- What information is missing?
- How would additional variables help?
- How can visualization mislead?
- How should uncertainty be communicated?

## AI Facilitation Guide
Prompt: Summarize the dataset, identify limitations, and suggest three scientific questions. Students should verify every claim against available evidence.

## Example AI Critique Exercise
Identify accurate statements, unsupported statements, and missing considerations. Focus on evidence rather than style.

## Common AI Hallucinations
- Inventing neuroscience findings
- Inferring disease status
- Overstating significance
- Ignoring sample size limitations

## Sample Student Artifact
A strong submission includes visualization, interpretation, discussion of limitations, AI critique, and reproducibility notes.

## TA Notes
Encourage evidence-based reasoning. Students should justify conclusions using the actual dataset rather than intuition.

## Troubleshooting Notes
If students struggle with interpretation, separate observation from explanation and revisit dataset limitations.

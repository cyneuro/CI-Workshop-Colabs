## ==============================================================================
## MODULE 1: CYBERINFRASTRUCTURE AUDITING & SCHEMA-VALIDATED DATA INGESTION
## ==============================================================================
**Mechanics:** Scientific pipelines must first establish trust in their computing container. We use underlying shell commands (`!`) to audit memory and processing footprints before pulling remote assets from open repositories.

```python
print("-------------------------------------------------------------------------")
print("[STAGE 1]: Auditing Underlying Host Cyberinfrastructure & Container Specs...")
print("-------------------------------------------------------------------------")

# Step 1.1: Programmatically extract hardware footprints from the Linux environment
import subprocess
import sys
import os
import json
import numpy as np
import matplotlib.pyplot as plt

try:
    cpu_info = subprocess.check_output("lscpu | grep 'Model name'", shell=True).decode().strip()
except:
    cpu_info = "Standard Virtual CPU Thread"
    
try:
    mem_info = subprocess.check_output("free -m | grep Mem", shell=True).decode().strip().split()
    total_ram = mem_info[1]
except:
    total_ram = "Unknown Local"

print(f"[CI AUDIT] Compute Node Processor: {cpu_info}")
print(f"[CI AUDIT] Volatile Memory Cache:  {total_ram} MB RAM Allocated")

print("\n-------------------------------------------------------------------------")
print("[STAGE 2]: Generating Synthetic Schema-Compliant Neural Telemetry Matrix...")
print("-------------------------------------------------------------------------")

def download_and_validate_dataset(subject_id="SUBJ-019"):
    """
    Simulates an automated API ingestion engine pulling from an open neuro-repository.
    Enforces a strict metadata schema gate: structural dimensions must be validated.
    """
    np.random.seed(42) 
    time_points = 500
    channels = 32
    
    raw_signal = np.random.normal(0.0, 1.0, (channels, time_points))
    t = np.linspace(0, 10, time_points)
    true_trajectory_x = np.sin(t) * 2.0
    true_trajectory_y = np.cos(t) * 1.5
    
    for i in range(5):
        raw_signal[i, :] += np.sin(t) * 3.5
        raw_signal[i+5, :] += np.cos(t) * 2.5
        
    dataset = {
        "metadata": {
            "dataset_origin": "DANDI_Archive_Synthetic_Mirror",
            "subject_identifier": subject_id,
            "sampling_rate_hz": 100,
            "channel_count": channels,
            "matrix_shape": raw_signal.shape
        },
        "data": {
            "raw_voltage_matrix": raw_signal.tolist(),
            "target_kinematics_x": true_trajectory_x.tolist(),
            "target_kinematics_y": true_trajectory_y.tolist()
        }
    }
    
    assert len(dataset["data"]["raw_voltage_matrix"]) == dataset["metadata"]["channel_count"], "CRITICAL: Schema Mismatch!"
    print(f"[SUCCESS]: Fetched and validated data stream for {subject_id}. Shape: {dataset['metadata']['matrix_shape']}")
    return dataset

neural_payload = download_and_validate_dataset()
```

## ==============================================================================
## MODULE 2: REUSABLE PRE-PROCESSING PIPELINE NODE
## ==============================================================================
### STUDENT AI PROMPT CHALLENGE:
Copy and paste the exact text inside the block quotation below into your AI assistant:

> "Write a modular python processing function named 'extract_neural_features(dataset_payload)' that performs downstream data preparation on our ingested neurophysiology matrix. The function must complete these configuration steps:
> 1. Extract the raw voltage matrix, sampling rate, and target kinematics from the dataset dict and convert them into standard 1D and 2D numpy arrays.
> 2. Apply a clean data reduction step: for each time index, average the voltage signal across the first 5 channels to isolate the principal motor cortical feature vector (call this feature_array_x). Average channels 5 to 10 to isolate the secondary feature vector (call this feature_array_y).
> 3. Return a clean python dictionary containing: 'feat_x': feature_array_x, 'feat_y': feature_array_y, 'true_x': target_kinematics_x, 'true_y': target_kinematics_y.
> Ensure arrays are properly aligned along the temporal axis."

```python
# STUDENT ZONE: PASTE THE GENERATED PRE-PROCESSING FUNCTIONS FROM YOUR AI ARCHITECT HERE:
def extract_neural_features(dataset_payload):
    raw_matrix = np.array(dataset_payload["data"]["raw_voltage_matrix"])
    true_x = np.array(dataset_payload["data"]["target_kinematics_x"])
    true_y = np.array(dataset_payload["data"]["target_kinematics_y"])
    
    feature_array_x = np.mean(raw_matrix[0:5, :], axis=0)
    feature_array_y = np.mean(raw_matrix[5:10, :], axis=0)
    
    processed_features = {
        "feat_x": feature_array_x,
        "feat_y": feature_array_y,
        "true_x": true_x,
        "true_y": true_y
    }
    return processed_features

# Execute Pre-processing Engine Node
pipeline_features = extract_neural_features(neural_payload)
print(f"[MODULE 2 SUCCESS]: Feature vectors isolated. Extracted dimension: {pipeline_features['feat_x'].shape}")
```

## ==============================================================================
## MODULE 3: THE COMPUTATIONAL ENGINE (INTENT DECODER CORE)
## ==============================================================================
### STUDENT AI PROMPT CHALLENGE:
Copy and paste the exact text inside the block quotation below into your AI assistant:

> "Write a mathematical computational decoding function named 'decode_motor_intent(feature_dictionary, scaling_factor)' to simulate a brain-machine interface tracking loop. The function must accept the feature dictionary emitted from our pre-processing module and execute this logic:
> 1. Extract 'feat_x' and 'feat_y' vectors.
> 2. Compute a smoothed intent prediction by multiplying the raw feature arrays by the 'scaling_factor' scalar parameter. To simulate biophysical feedback delays, apply a 3-point rolling moving average using np.convolve(vector, np.ones(3)/3, mode='same') across both generated prediction vectors.
> 3. Return a tuple containing two numpy arrays: (decoded_trajectory_x, decoded_trajectory_y)."

```python
# STUDENT ZONE: PASTE THE GENERATED MODEL MATHEMATICS COMPUTATION FUNCTIONS HERE:
def decode_motor_intent(feature_dictionary, scaling_factor=0.50):
    fx = feature_dictionary["feat_x"]
    fy = feature_dictionary["feat_y"]
    
    raw_pred_x = fx * scaling_factor
    raw_pred_y = fy * scaling_factor
    
    kernel = np.ones(3) / 3.0
    decoded_trajectory_x = np.convolve(raw_pred_x, kernel, mode='same')
    decoded_trajectory_y = np.convolve(raw_pred_y, kernel, mode='same')
    
    return decoded_trajectory_x, decoded_trajectory_y

# Set Baseline System Parameter Valve (The Lever)
DECODER_GAIN = 0.52

# Execute Computational Decoder Engine
pred_x, pred_y = decode_motor_intent(pipeline_features, scaling_factor=DECODER_GAIN)
print("[MODULE 3 SUCCESS]: Kinematic decoding matrices populated safely in local runtime memory cache.")
```

## MODULE 4: POST-PROCESSING ANALYTICS & METRIC TELEMETRY SHIELD
## ==============================================================================
### STUDENT AI PROMPT CHALLENGE:
Copy and paste the exact text inside the block quotation below into your AI assistant:

> "Write a comprehensive matplotlib rendering dashboard function named 'deploy_scientific_dashboard(features, pred_x, pred_y)'. The layout must feature a 2-panel configuration inside a single viewport frame:
> 1. Panel 1 (Top Coordinate Viewport): Display the true_x trajectory line against the decoded pred_x trajectory line across the time steps.
> 2. Panel 2 (Bottom Status Card Area): Remove coordinate borders. Calculate the Mean Squared Error (MSE) between true_x and pred_x. 
>    - If MSE is less than 0.75, draw a green text scorecard saying 'COMPLIANCE GATE STATUS: PASSED (OPTIMAL WORKFLOW)'.
>    - If MSE is 0.75 or greater, draw a red text scorecard saying 'COMPLIANCE GATE STATUS: FAILED (ACCURACY DRIFT ERROR)'.
> Ensure tight axis margins and clear typographic grid labels."

```python
# STUDENT ZONE: PASTE THE GENERATED POST-PROCESSING GRAPHICS INTERFACE DESIGN DIRECTLY HERE:
def deploy_scientific_dashboard(features, pred_x, pred_y):
    true_x = features["true_x"]
    mse_error = np.mean((true_x - pred_x) ** 2)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 6), gridspec_kw={'height_ratios': [3, 1]})
    
    ax1.plot(true_x, label='True Kinematic Path (Target)', color='#2ca02c', linewidth=3)
    ax1.plot(pred_x, label='Decoded Intention Path (Model Output)', color='#d62728', linewidth=2, linestyle='--')
    ax1.set_title("Automated Workflow Verification Dashboard — Telemetry Layer", fontsize=12, fontweight='bold')
    ax1.set_xlabel("Pipeline Operational Time Arrays (ms)", fontsize=10)
    ax1.set_ylabel("Spatial Position Coordinate Delta", fontsize=10)
    ax1.legend(loc='upper right')
    ax1.grid(True, linestyle=':', alpha=0.6)
    
    ax2.axis('off')
    if mse_error < 0.75:
        bg, border_color, message = '#d4edda', '#155724', f"COMPLIANCE GATE STATUS: PASSED (OPTIMAL DATA PIPELINE) \nMean Squared Error Analytics: {mse_error:.4f}"
    else:
        bg, border_color, message = '#f8d7da', '#721c24', f"COMPLIANCE GATE STATUS: FAILED (ACCURACY DRIFT OUTSIDE BOUNDS) \nMean Squared Error Analytics: {mse_error:.4f}"
        
    ax2.text(0.5, 0.5, message, size=11, weight='bold', color=border_color, ha="center", va="center",
             bbox=dict(boxstyle="round,pad=0.6", facecolor=bg, edgecolor=border_color, lw=2))
    
    plt.tight_layout()
    plt.show()
    return mse_error

system_mse = deploy_scientific_dashboard(pipeline_features, pred_x, pred_y)
```

**Mechanics:** Data without structured origin paths is unscientific noise. Here, we programmatically capture the hardware footprints, software hashes, and hyperparameter configurations to generate a reproducible data sidecar.

```python
print("-------------------------------------------------------------------------")
print("[PROVENANCE ENFORCEMENT]: Packaging System Architecture Sidecar Manifest...")
print("-------------------------------------------------------------------------\n")

def export_workflow_provenance(subject_meta, decoder_gain, metric_mse):
    provenance_manifest = {
        "workflow_engine_metadata": {
            "execution_platform": "Google Colab VM",
            "operating_system_posix": os.name,
            "python_runtime_version": sys.version.split()[0],
            "underlying_numerical_precision": str(np.float64)
        },
        "scientific_provenance": {
            "source_dataset_origin": subject_meta["dataset_origin"],
            "target_subject_id": subject_meta["subject_identifier"],
            "registered_channel_count": subject_meta["channel_count"]
        },
        "pipeline_hyperparameters": {
            "decoder_gain_value": decoder_gain,
            "convolution_smoothing_window_points": 3
        },
        "automated_validation_metrics": {
            "mean_squared_error_result": metric_mse,
            "reproducibility_clearance_status": "APPROVED" if metric_mse < 0.75 else "REJECTED"
        }
    }
    
    output_filename = "workflow_provenance_manifest.json"
    with open(output_filename, 'w') as json_file:
        json.dump(provenance_manifest, json_file, indent=4)
        
    print(f"[METADATA SUCCESS]: Saved compliance tracking record to: '{output_filename}'")
    print(json.dumps(provenance_manifest, indent=2))

export_workflow_provenance(neural_payload["metadata"], DECODER_GAIN, system_mse)
```

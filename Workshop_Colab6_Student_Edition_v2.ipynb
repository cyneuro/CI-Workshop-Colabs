## Activity 1: Explore Features

```python
import pandas as pd
url="https://raw.githubusercontent.com/OpenNeuroDatasets/ds000247/main/participants.tsv"
df=pd.read_csv(url,sep="\t")
df.head()
```

### Questions
- Which variables are available?
- Which variables might be useful for machine learning?
- What important information is missing?

## Activity 2: Feature Engineering

```python
df=df[df["age"]!="n/a"].copy()
df["age"]=pd.to_numeric(df["age"])
df[["age"]].describe()
```

### Reflection
How could additional behavioral, cognitive, EEG, MEG, or imaging variables improve analysis?

## Activity 3: Clustering Example

```python
from sklearn.cluster import KMeans
X=df[["age"]]
model=KMeans(n_clusters=2,random_state=42,n_init=10)
df["cluster"]=model.fit_predict(X)
df[["age","cluster"]].head()
```

### Interpretation Questions
- What do the clusters represent?
- What information is the algorithm using?
- What conclusions should not be drawn?

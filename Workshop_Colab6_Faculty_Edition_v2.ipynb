## Common Misconceptions
1. Machine learning discovers truth automatically.
2. Complex models are always better.
3. Clusters represent real biological groups.
4. AI explanations are trustworthy by default.

```python
import pandas as pd
from sklearn.cluster import KMeans
url="https://raw.githubusercontent.com/OpenNeuroDatasets/ds000247/main/participants.tsv"
df=pd.read_csv(url,sep="\t")
df=df[df["age"]!="n/a"].copy()
df["age"]=pd.to_numeric(df["age"])
```

## Activity 1: Feature Selection
Students evaluate which variables are available and what additional neuroscience variables would improve analysis.

### Faculty Solution
Current features are limited. Additional EEG, MEG, behavioral, cognitive, clinical, and longitudinal measures would strengthen analyses.

```python
X=df[["age"]]
model=KMeans(n_clusters=2,random_state=42,n_init=10)
df["cluster"]=model.fit_predict(X)
df[["participant_id","age","cluster"]].head()
```

## Facilitation Notes
Use clustering to demonstrate how features determine discoveries. The algorithm can only discover patterns available in the selected variables.

## Activity 2: Clustering Interpretation
Students examine cluster assignments and evaluate whether meaningful conclusions are justified.

### Faculty Solution
The clusters primarily reflect age differences because age is the only feature used.

## Validation Framework

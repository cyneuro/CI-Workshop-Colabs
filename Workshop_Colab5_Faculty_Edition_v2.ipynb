## Teaching Narrative
Students learn when workflows outgrow a laptop and require cloud, HPC, or national cyberinfrastructure resources. Emphasis is on workflow design decisions rather than specific technologies.

Colabs 1–4 established workflow foundations, repositories, analysis, and reproducibility. Colab 5 introduces workflow scaling and cyberinfrastructure.

## NSF CyberTraining Alignment

## Learning Outcomes
- Estimate workflow resource needs
- Compare laptop, cloud, and HPC environments
- Identify bottlenecks
- Justify infrastructure selection
- Design scalable workflows

## Instructor Background
Modern science increasingly relies on distributed resources. Workflow requirements are determined by data volume, compute intensity, collaboration needs, and reproducibility constraints.

## Common Misconceptions
1. Bigger datasets always require HPC.
2. Cloud and HPC are identical.
3. GPUs solve every scaling problem.
4. Scaling only concerns computation.

## Activity 1: Resource Estimation
Students estimate storage, memory, compute, and collaboration requirements.

## Facilitation Notes
Students should justify decisions using evidence. Focus on workflow characteristics rather than hardware preferences.

## Workflow Scaling Framework
Data Size → Compute Needs → Storage Needs → Collaboration Needs → Infrastructure Choice

## Cloud vs HPC Discussion
Cloud: elasticity, services, rapid provisioning. HPC: tightly coupled computation, large simulations, parallel processing.

## Cyberinfrastructure Case Studies
Discuss examples such as FABRIC, CloudLab, Chameleon, Open Science Grid, and neuroscience computing resources.

## Activity 2 Faculty Solution
Laptop: classroom datasets. Cloud VM: moderate-scale workflows. HPC: large analytics, simulations, or ML training. National CI: highly distributed projects.

## Strong Student Response
Justifies infrastructure decisions using resource estimates, scalability, reproducibility, and collaboration needs.

## Average Student Response
Identifies infrastructure categories but provides weak justification.

## Weak Student Response
Recommends infrastructure without considering workflow requirements.

## Bottleneck Analysis Framework
Consider storage, memory, compute, networking, software dependencies, and human workflows.

## Discussion Prompts
- When is a laptop enough?
- What creates bottlenecks?
- When should GPUs be used?
- What changes as collaborations grow?

## AI Facilitation Guide
Prompt: Redesign this workflow for a dataset 1000 times larger. Evaluate assumptions, tradeoffs, and missing considerations.

## Common AI Hallucinations
- Infinite cloud resources
- Ignoring data transfer costs
- Assuming GPUs solve all problems
- Ignoring reproducibility needs

## Sample Student Artifact
Resource estimate, architecture diagram, bottleneck analysis, scaling strategy, and AI critique.

## TA Notes
Guide students toward tradeoff analysis instead of vendor-specific discussions.

## Troubleshooting Notes
Use familiar examples such as smartphones, laptops, and online storage services when students struggle with scale.

## Online Delivery Notes
Assign groups different workflow scales and compare infrastructure recommendations.

## Stretch Challenge

## Transition to Semester Course

## Faculty Takeaways

## Instructor Background
Neuroinformatics provides methods for organizing, preserving, sharing, discovering, and reusing neuroscience data. Discuss FAIR data practices and community standards.

## OpenNeuro Concepts
OpenNeuro provides open neuroscience datasets organized using BIDS. Discuss dataset reuse, transparency, and reproducibility.

## BIDS Concepts
Key artifacts: dataset_description.json, README, participants.tsv, modality-specific data files, derivatives, and metadata records.

## Common Misconceptions
1. Data files are sufficient without metadata.
2. Repository submission guarantees scientific quality.
3. FAIR means error-free.
4. Provenance and metadata are identical.

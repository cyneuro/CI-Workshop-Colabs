# Cloud-to-Edge Scientific Workflows and Embedded Analytics

# FACULTY EDITION v3

## Teaching Narrative
Scientific workflows increasingly span sensors, edge devices, cloud resources, HPC systems, AI services, and users. Focus on workflow architecture and placement decisions.

## Workshop Context
Colab 1 Workflow Foundations; Colab 2 Repositories; Colab 3 Analysis; Colab 4 Reproducibility; Colab 5 Scaling; Colab 6 ML; Colab 7 Cloud-to-Edge; Colab 8 Validation and Trust.

## NSF CyberTraining Alignment
Distributed computing, edge computing, cyberinfrastructure, AI workflows, embedded analytics, and data-intensive science.

## Learning Outcomes
Explain cloud, edge, and HPC roles; design cloud-to-edge workflows; evaluate latency/bandwidth tradeoffs; justify workflow placement decisions.

## Instructor Delivery Guide
0–10 Motivation; 10–25 Sensor workflows; 25–40 Edge analytics; 40–55 Cloud/HPC integration; 55–70 AI workflow design; 70–80 Reflection.

## Activity Timing Guide
Sensor Workflow Analysis 15 min; Architecture Design 15 min; Resource Placement 15 min; AI Critique 15 min.

## Instructor Background
Modern workflows span sensors → edge → cloud → HPC → AI → users. Workflow architecture is a scientific design decision.

## Common Misconceptions
Everything belongs in the cloud; edge replaces cloud; HPC equals cloud; AI must run centrally; more computation is always better.

## Workflow Placement Framework
Edge: acquisition, filtering, compression, alerts. Cloud: storage, dashboards. HPC: training, simulation, large-scale analytics.

## Faculty Solution
Students should justify placement using latency, bandwidth, storage, compute, scalability, provenance, and reproducibility considerations.

## Neuroscience Case Study
Mobile EEG workflow: Sensor → Wearable → Edge Processing → Cloud Storage → ML Analysis → Dashboard.

## Case Study Solution
Edge: artifact detection and filtering. Cloud: storage and collaboration. HPC: cohort-scale analysis and model training.

## Wearable Monitoring Case Study
Evaluate local decision making, synchronization, AI integration, and scalability.

## Strong / Average / Weak Responses
Strong: justified architecture. Average: identifies layers. Weak: places everything in one location.

## Environmental Monitoring Case Study
Discuss whether all sensor readings should be transmitted and what computations belong locally.

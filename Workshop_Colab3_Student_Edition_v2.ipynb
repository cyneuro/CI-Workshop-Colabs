## Activity 1: Explore a Dataset

```python
import pandas as pd
url="https://raw.githubusercontent.com/OpenNeuroDatasets/ds000247/main/participants.tsv"
df=pd.read_csv(url,sep="\t")
df.head()
```

```python
df.columns
```

## Activity 2: Descriptive Statistics

```python
df=df[df["age"]!="n/a"].copy()
df["age"]=pd.to_numeric(df["age"])
df["age"].describe()
```

### Reflection
What do these statistics reveal? What do they fail to reveal?

## Activity 3: Visualization

```python
import matplotlib.pyplot as plt
plt.hist(df["age"])
plt.title("Age Distribution")
plt.show()
```

- What patterns are visible?
- What additional data would improve interpretation?
- What conclusions should not be drawn?

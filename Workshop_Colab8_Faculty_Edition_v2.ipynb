## Instructor Background
Scientific workflows must be reproducible, transparent, auditable, and trustworthy. Validation and benchmarking support confidence in results.

## Common Misconceptions
Reproducible means correct; FAIR guarantees quality; AI outputs are automatically trustworthy; benchmarking is only about speed.

## Validation Framework
Validate inputs, metadata, code, parameters, outputs, provenance, assumptions, and interpretations.

## Faculty Solution
Students should distinguish correctness, reproducibility, validation, and trustworthiness as related but different concepts.

## Benchmarking Methodology
Evaluate runtime, memory use, portability, scalability, robustness, and reproducibility across environments.

## FAIR Assessment Framework
Findable, Accessible, Interoperable, Reusable. Require evidence supporting each assessment.

## Workflow Audit Template
Dataset source, version, code version, software versions, parameters, AI usage, outputs, documentation, and provenance.

## AI Trustworthiness Checklist
Transparency, validation, reproducibility, limitations, uncertainty reporting, human oversight, and bias awareness.

## Strong / Average / Weak Responses
Strong: evidence-based audit. Average: checklist only. Weak: assumes workflow is trustworthy without validation.

## Reproducibility Failure Case Studies
Missing versions, undocumented preprocessing, unavailable code, hidden parameters, and undocumented AI assistance.

## Discussion Prompts
Can a reproducible workflow still be wrong? What makes AI trustworthy? How should workflows be audited?

## AI Facilitation Guide
Prompt: Evaluate a workflow for trustworthiness and identify risks. Students critique assumptions and missing evidence.

## Common AI Hallucinations
Claiming validation without evidence, assuming FAIR compliance, inventing documentation, overstating reliability.

## Sample Student Artifact
Workflow audit report, FAIR assessment, benchmarking summary, AI critique, and reproducibility review.

## TA Notes
Emphasize evidence-based evaluation and audit thinking rather than checklist completion.

## Troubleshooting Notes
Use simple examples to separate reproducibility, correctness, and trustworthiness when concepts become confused.

## Online Delivery Notes
